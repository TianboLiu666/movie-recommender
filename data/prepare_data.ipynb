# Merge

```python
final_movies = movies.merge(plot, on=['Name', 'year'], how='left')#.head()#.info()
```

```python
final_movies.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
type(final_movies.fillna('')['Description'])
```

```python
corpus = final_movies['Name'] #+ final_movies.fillna('')['Genres'] #+  final_movies.fillna('')['Keywords']
```

```python
tfidf_vectorizer = TfidfVectorizer()


tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
```

```python
import numpy as np
np.ones((2,3)).tolist()
```

```python
tfidf_matrix.todense()
```

```python
pd.Series(list(tfidf_matrix.toarray()))
```

```python
# final_movies['tfidf_embedding'] = pd.Series(list(tfidf_matrix.toarray()))
final_movies['tfidf_embedding'] = pd.Series(np.array(tfidf_matrix.todense()).tolist())
```

```python
type(final_movies['tfidf_embedding'][0][0])
```

```python
final_movies.head()
```

```python
final_movies[final_movies['year'] > 2018].head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import time

# Record start time
start = time.time()

# Compute cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Print cosine similarity matrix
print(cosine_sim)

# Print time taken
print("Time taken: %s seconds" % (time.time() - start))
```
